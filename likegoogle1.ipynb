```python
df=pd.read_csv('countries.csv', error_bad_lines=False)


prod = pd.DataFrame({'latitude': [], 'longitude': [], 'country': [],'Number of cases':[]})

#print(df)
for i in range(0,len(df)):
    for ii in range(0,len(label_sort_df_confineT_20)):
        if(df.iloc[i]['name']==value_sort_df_confineT_20.columns[ii]):
           # print(df.iloc[i]['latitude'],df.iloc[i]['longitude'],df.iloc[i]['name'],value_sort_df_confineT_20.iloc[0][value_sort_df_confineT_20.columns[ii]])
            prod = prod.append({'latitude': df.iloc[i]['latitude'], 'longitude' : df.iloc[i]['longitude'], 'country' : df.iloc[i]['name'],'Number of cases' :value_sort_df_confineT_20.iloc[0][value_sort_df_confineT_20.columns[ii]]},ignore_index=True)
print(world_geo)
print(prod)            

prod.replace({
        'US':'United States of America'},
        inplace=True)
       


#3/20/20
import datetime
date=datetime.datetime.strptime(first_column, '%m/%d/%y').strftime('%d-%m-%y')

print(date)

#m1 = folium.Map(location=[0, 0], zoom_start=1)
m1 = folium.Map(location=[0, 0], zoom_start=1)

#print(state_data)
folium.Choropleth(
    geo_data=world_geo,
    name='choropleth',
    data=prod,
    columns=['country','Number of cases'],
    key_on='feature.properties.name',
    #fill_color='YlGn',
    fill_color='PRGn',
    fill_opacity=0.8,
    line_opacity=0.5,
    legend_name='Number of confirmed cases'
).add_to(m1)

folium.LayerControl().add_to(m1)

m1.save(date+'goolge_confirmed.html')
m1
```
